```python
import pandas as pd
```

```python
df = pd.read_csv('ind_nifty50list.csv')
```

```python
# Display top 4 rows
df.head(4)
```

```python
# Display 4th row with head and tail
df.head(4).tail(1)
```

```python
# Display 4th row with iloc func, iloc uses only row index and column index
df.iloc[[3]]
```

```python
# Display 4th row and specific columns only
df.iloc[[3],[0,4]]
```

```python
# Display 2nd,4th rows and 1st & 4th column
df.iloc[[1,3],[0,3]]
```

```python
# loc function, if we use just row index and NO column it display all columns
df.loc[3]
```

```python
# Using loc func, this uses row index and column name
df.loc[3,['Company Name']]
```

```python
df.loc[3,['Company Name','ISIN Code']]
```
